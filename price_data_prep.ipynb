```python
import pandas as pd
```

```python
file = "RDC_InventoryCoreMetrics_Zip_Hist.csv"

# Read our Kickstarter data into pandas
df = pd.read_csv(file)
df.head()
```

```python
df.columns
```

```python
zip_data = df.loc[:, ["Month", "ZipCode", "ZipName","Median Listing Price", 
                                    "Active Listing Count ", "Avg Listing Price"]]
zip_data.head()
```

```python
zip_data[['City','State']] = zip_data.ZipName.str.split(",",expand=True,)
zip_data.head()
```

```python
zip_data['State'] = zip_data['State'].str.replace(' ', '')
zip_data.head()
```

```python
zip_data[['Month1','Day', 'Year']] = zip_data.Month.str.split("/",expand=True,)
zip_data.head()
```

```python
agg = {'Active Listing Count ': ['sum'], 'Avg Listing Price': ['mean']}
```

```python
State_data_df = zip_data.groupby(['State', 'Year']).agg(agg).reset_index()
State_data_df.head(5)
```

```python
State_data_df.columns = ['State', 'Year', 'Active Listing Count', 'Avg Listing Price']
State_data_df.head()
```

```python
State_data_df = State_data_df.sort_values(['Year', 'State'])
State_data_df.head(20)
```

```python
State_data_json = State_data_df.to_json(orient='records')
```

```python
State_data_json_final = eval(State_data_json)
```
